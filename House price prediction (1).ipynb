```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x1 = data.drop(['id','price','date','waterfront','view','condition','sqft_above','sqft_basement','yr_built','yr_renovated','lat','long','sqft_living15','sqft_lot15'],axis=1)
```

```python
conv_dates = [1 if values == 2014 else 0 for values in data.date ]
data['date'] = conv_dates
train1 = data.drop(['id', 'price'],axis=1)
```

```python
y1 = data['price']
```

```python
x1.head()
```

```python
y1.head()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x1,y1,test_size = 0.10,random_state=2)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```
